```python
r = sr.Recognizer()
with sr.Microphone() as source:
    print("Speak something...")
    r.adjust_for_ambient_noise(source) 
    audio = r.listen(source, timeout=3, phrase_time_limit=10)
```

```python
# Converting the audio to text
try:
    text = r.recognize_google(audio)
    print("You said:", text)
    
except sr.UnknownValueError:
    print("❌ Could not understand audio")

except sr.RequestError:
    print("❌ API unavailable / Internet issue")
    
```

Using Regular Expressions

1. Extracting only the mobile numbers using Regular Expression

```python
sent = 'Hello Nitesh, nice meeting with you.Lets be in touch please note down this mobile number which is 8741256412.If this number not working will give you another alternate number being 7845612394'
import re
re.findall('[0-9]{10}', sent)
```

2. Now hiding the mobile nos with stars and printing the sentence

```python
re.sub('[0-9]{10}', '**********', sent)
```

3. In the below sentence extracting the mail ID,phone number,date

```python
sent2 = 'Hey Mark,please create a GMAIL account for me.Keep my mailID as stephen.king-1407@gmail.com and password as stephen_king666.My mobile number is 555-777-122 and date of birth is 13/12/1989.'
pattern = '[0-9a-zA-Z.-]+@[0-9a-zA-Z.-]+'
re.findall(pattern,sent2)
```

```python
pattern1 = '[0-9a-z]+_[0-9a-z]+'
re.findall(pattern1,sent2)
```

```python
pattern2 = '[0-9-]{11}'
re.findall(pattern2,sent2)
```

```python
pattern3 = '[0-9]{1,2}/[0-9]{1,2}/[0-9]{4}'
re.findall(pattern3,sent2)
```

Hiding the password from the sentence

```python
re.sub(pattern1,"",sent2)
```
